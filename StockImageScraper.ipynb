```python
import os
import requests
import pandas as pd
from selenium.webdriver.chrome.service import Service
import chromedriver_binary
from tqdm import tqdm

from selenium import webdriver
from bs4 import BeautifulSoup

link = 'https://stock-pictures.netlify.app'
```

```python
# Create a Service object with the chromedriver path
service = Service(chromedriver_binary.chromedriver_filename)

# Pass the service to the Chrome driver
driver = webdriver.Chrome(service=service)
```

```python
driver.get(link)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
print(len(soup.find_all('img')))
```

```python
data = []
i = 1
for sp in soup.find_all('div', class_ = 'container'):
    # print(sp) 
    # print('-'*20)

    if('gif' not in sp.find('img').get('src')):
        link = sp.find('img').get('src')
        tags = list(set(sp.find('div', class_ = 'tags').text[7:].split(',')))
        likes = int(sp.find('div', class_ = 'likes-comments').find_all('span')[0].text[:-6])
        comments = int(sp.find('div', class_ = 'likes-comments').find_all('span')[1].text[:-9])

        data.append([link, ' '.join(tags), likes, comments])
        
#data
#print(len(links))
```

```python
df = pd.DataFrame(data, columns=['Img Link', 'Tags', 'Likes', 'Comments'])
```

2. Downloading all Images

```python
path = []

for link in tqdm(df['Img Link']):
    res = requests.get(link)
    
    # Define the image directory
    img_dir = r"C:\Project ideas\Web Scraping\Image Scraper\Imgs"
    
    # Ensure the directory exists
    os.makedirs(img_dir, exist_ok=True)
    
    # Define the image file path
    file_path = os.path.join(img_dir, link.split('/')[-1])
    
    # Open and write image content
    with open(file_path, "wb") as fd:
        fd.write(res.content)
    
    path.append(file_path)
```
